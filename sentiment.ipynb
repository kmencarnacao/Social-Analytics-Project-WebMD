```python
import nltk
import pandas as pd
```

```python
from textblob import TextBlob
```

```python
webmd = pd.read_csv('webmd-1.csv')
webmd
```

```python
#this represents the length of the reviews
webmd["reviews_length"] = webmd.Reviews.apply(lambda x: len(x))
webmd[["Reviews", "reviews_length"]]
```

```python
#this represents the token and the tagged tokens of the reviews
webmd["tokens"] = webmd.Reviews.apply(lambda x: nltk.word_tokenize(x))
webmd["tagged_tokens"] = webmd.tokens.apply(lambda x: nltk.pos_tag(x))
webmd[["Reviews", "tokens", "tagged_tokens"]].head()
```

```python
#Reviews sentiment
webmd["polarity"] = webmd.Reviews.apply(lambda x: TextBlob(x).sentiment.polarity)
webmd["subjectivity"] = webmd.Reviews.apply(lambda x: TextBlob(x).sentiment.subjectivity)
webmd[["Reviews", "polarity", "subjectivity"]]
```

```python
webmd[(webmd.polarity > 0.7) | (webmd.polarity < -0.7)][["Reviews", "polarity"]]
```
